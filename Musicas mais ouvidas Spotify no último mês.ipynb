```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
import plotly.express as px

py.init_notebook_mode(connected=True)
```

```python
data1 = pd.read_csv('regional-global-weekly-2022-06-02.csv')
data1['data'] = '06-02-2022'
data2 = pd.read_csv('regional-global-weekly-2022-06-09.csv')
data2['data'] = '06-09-2022'
data3 = pd.read_csv('regional-global-weekly-2022-06-16.csv')
data3['data'] = '06-16-2022'
data4 = pd.read_csv('regional-global-weekly-2022-06-23.csv')
data4['data'] = '06-23-2022'

data1.head()
```

```python
data1 = data1.drop(['uri', 'previous_rank', 'weeks_on_chart'], axis = 1)
data2 = data2.drop(['uri', 'previous_rank', 'weeks_on_chart'], axis = 1)
data3 = data3.drop(['uri', 'previous_rank', 'weeks_on_chart'], axis = 1)
data4 = data4.drop(['uri', 'previous_rank'], axis = 1)
```

```python
data1.head()
```

```python
df = pd.merge(data1, data2, how = 'outer')
df = pd.merge(df, data3, how = 'outer')
df = pd.merge(df, data4, how = 'outer')

df.info()
```

```python
datas = df['data']
```
